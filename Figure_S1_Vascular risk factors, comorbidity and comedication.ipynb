```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
path = './figure_S3_data.xlsx'

df = pd.read_excel(path)

df.info()
```

```python
df
```

```python
df_user = df[df.statin == 'user']['prevalence'].values
df_covariate = df[df.statin == 'user']['covariate'].values
df_nonuser = df[df.statin == 'nonuser']['prevalence'].values
#covariates = list(set(df.covariate))

frame = pd.DataFrame({'covariate': df_covariate, 'statin_users': df_user, 'non_users': df_nonuser})

frame
```

```python
frame.covariate = ['≥70 years', 'Men', 'BMI≥25', 'Alcohol abuser', 'Current smoker', 'Chronic gastritis',
                   'Gastroesophageal reflux', 'Inflammatory bowel disease', 'Irritable bowel syndrome',
                   'Constipation', 'Anorectal pathology', 'Complicated upper GI disorder', 
                   'Non-complicated upper GI disorder', 'Dyspepsia', 'Lower GI bleeding', 'Gout', 'Hyperuricemia',
                   'Oral anticoagulants', 'Glucose-lowering drugs', 'Low-dose aspirin', 'NSAIDs', 
                   'Other lipid-lowering drugs', 'Antihypertensives', 'Diabetes', 'Hypertension',
                   'Peripheral artery disease', 'Acute myocardial infarction', 'Angina pectoris', 'Stroke',
                   'Transient ischemic attack']
frame
```

```python
frame.sort_values('statin_users', ascending=False, inplace=True)
frame
```

```python
sns.set_theme()
labels = ['Current users', 'Non-users']

fig = plt.figure(figsize=(30,15))
plt.style.use('default')

plt.plot(frame.covariate,frame.statin_users, c='orangered', linewidth=3)
plt.scatter(frame.covariate,frame.statin_users, c='orangered', s=(frame.statin_users)*10)
plt.fill_between(frame.covariate,frame.statin_users, where=frame.statin_users>0, 
                 interpolate=True, color='coral', alpha=0.1)

plt.fill_between(frame.covariate,frame.non_users, where=frame.non_users>0, 
                 interpolate=True, color='darkgoldenrod', alpha=0.1)
plt.plot(frame.covariate,frame.non_users, c='goldenrod', linewidth=3)
plt.scatter(frame.covariate,frame.non_users, c='goldenrod', s=(frame.non_users)*10)

plt.legend(labels, loc=5, prop={'size':25})
plt.xticks(rotation=90, fontsize=25, fontweight='medium')
plt.yticks(fontsize=25, fontweight='medium')
plt.ylim(0, 80)
plt.ylabel('Prevalence (%)', loc='center', fontsize=30, fontweight='bold')
plt.savefig(fname='./Figure_S3_burden_of_disease.pdf', bbox_inches='tight', dpi=300)
plt.show()
```
